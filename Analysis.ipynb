Были проведения сравнения бинарайзеров Hinton, Concrete,HardKuma,  а также их замена обычной биномиальной случайной величиной с вероятностями 20%, 50%, 80%, 90% и 99%.

По результатам выяснилось что наименьшее значение функции потерь достигается при использовании Hinton, а наибольшое значение F1 - при биномиальной с.в. с вероятностью 99%.

 Биномиальную с.в. можно использовать для замени бинарайзеров только с p=90%+,потому что только они давали близкое к бинарайзерам значение функции потерь, но все равно проигрывали

```python
import pandas as pd
```

```python
def get_iter(x, t = 'train'):
    if t in x:
        return x.split('/')[0]
    
def get_loss(x, t = 'train'):
    if t in x:
        for i in x.split(' | '):            
            if 'loss=' in i:
                return float(i.split('loss=')[1])
def get_f1(x, t = 'train'):
    if t in x:
        for i in x.split(' | '):
            if 'f1_score=' in i:
                return float(i.split('f1_score=')[1])           
```

```python
Hinton = pd.read_csv('Hinton/log.txt',header = None)
```

```python
hinton_loss_train = [get_loss(x, t = 'train') for x in Hinton[0] if 'train' in x ]
hinton_loss_valid = [get_loss(x, t = 'valid') for x in Hinton[0] if 'valid' in x]
hinton_f1_train = [get_f1(x, t = 'train') for x in Hinton[0] if 'train' in x ]
hinton_f1_valid = [get_f1(x, t = 'valid') for x in Hinton[0] if 'valid' in x]
```

```python
Concrete = pd.read_csv('Concrete/log.txt',header = None)
concrete_loss_train = [get_loss(x, t = 'train') for x in Concrete[0] if 'train' in x ]
concrete_loss_valid = [get_loss(x, t = 'valid') for x in Concrete[0] if 'valid' in x]
concrete_f1_train = [get_f1(x, t = 'train') for x in Concrete[0] if 'train' in x ]
concrete_f1_valid = [get_f1(x, t = 'valid') for x in Concrete[0] if 'valid' in x]
```

```python
HardKuma = pd.read_csv('HardKuma/log.txt',header = None)
HardKuma_loss_train = [get_loss(x, t = 'train') for x in HardKuma[0] if 'train' in x ]
HardKuma_loss_valid = [get_loss(x, t = 'valid') for x in HardKuma[0] if 'valid' in x]
HardKuma_f1_train = [get_f1(x, t = 'train') for x in HardKuma[0] if 'train' in x ]
HardKuma_f1_valid = [get_f1(x, t = 'valid') for x in HardKuma[0] if 'valid' in x]
```

```python
b20 = pd.read_csv('b20/log.txt',header = None)
b20_loss_train = [get_loss(x, t = 'train') for x in b20[0] if 'train' in x ]
b20_loss_valid = [get_loss(x, t = 'valid') for x in b20[0] if 'valid' in x]
b20_f1_train = [get_f1(x, t = 'train') for x in b20[0] if 'train' in x ]
b20_f1_valid = [get_f1(x, t = 'valid') for x in b20[0] if 'valid' in x]
```

```python
b50 = pd.read_csv('b50/log.txt',header = None)
b50_loss_train = [get_loss(x, t = 'train') for x in b50[0] if 'train' in x ]
b50_loss_valid = [get_loss(x, t = 'valid') for x in b50[0] if 'valid' in x]
b50_f1_train = [get_f1(x, t = 'train') for x in b50[0] if 'train' in x ]
b50_f1_valid = [get_f1(x, t = 'valid') for x in b50[0] if 'valid' in x]
```

```python
b80 = pd.read_csv('b80/log.txt',header = None)
b80_loss_train = [get_loss(x, t = 'train') for x in b80[0] if 'train' in x ]
b80_loss_valid = [get_loss(x, t = 'valid') for x in b80[0] if 'valid' in x]
b80_f1_train = [get_f1(x, t = 'train') for x in b80[0] if 'train' in x ]
b80_f1_valid = [get_f1(x, t = 'valid') for x in b80[0] if 'valid' in x]
```

```python
b90 = pd.read_csv('b90/log.txt',header = None)
b90_loss_train = [get_loss(x, t = 'train') for x in b90[0] if 'train' in x ]
b90_loss_valid = [get_loss(x, t = 'valid') for x in b90[0] if 'valid' in x]
b90_f1_train = [get_f1(x, t = 'train') for x in b90[0] if 'train' in x ]
b90_f1_valid = [get_f1(x, t = 'valid') for x in b90[0] if 'valid' in x]

b99 = pd.read_csv('b99/log.txt',header = None)
b99_loss_train = [get_loss(x, t = 'train') for x in b99[0] if 'train' in x ]
b99_loss_valid = [get_loss(x, t = 'valid') for x in b99[0] if 'valid' in x]
b99_f1_train = [get_f1(x, t = 'train') for x in b99[0] if 'train' in x ]
b99_f1_valid = [get_f1(x, t = 'valid') for x in b99[0] if 'valid' in x]
```

```python
train_loss = pd.DataFrame(
                        [hinton_loss_train,
                           HardKuma_loss_train,
                           concrete_loss_train,
                           b20_loss_train,
                           b50_loss_train,
                           b80_loss_train,
                           b90_loss_train,
                             b99_loss_train,
                            ]
    ).T
train_loss.columns = ['hinton','hard-kuma','concrete',
                      'binom 20%','binom 50%','binom 80%','binom 90%','binom 99%']

valid_loss = pd.DataFrame(
                        [hinton_loss_valid,
                           HardKuma_loss_valid,
                           concrete_loss_valid,
                           b20_loss_valid,
                           b50_loss_valid,
                           b80_loss_valid,
                           b90_loss_valid,      
                         b99_loss_train
                            ]
    ).T
valid_loss.columns = ['hinton','hard-kuma','concrete',
                      'binom 20%','binom 50%','binom 80%','binom 90%','binom 99%']

train_f1 = pd.DataFrame(
                        [hinton_f1_train,
                           HardKuma_f1_train,
                           concrete_f1_train,
                           b20_f1_train,
                           b50_f1_train,
                           b80_f1_train,
                           b90_f1_train,
                         b99_loss_train
                            ]
    ).T
train_f1.columns = ['hinton','hard-kuma','concrete',
                      'binom 20%','binom 50%','binom 80%','binom 90%','binom 99%']
```
